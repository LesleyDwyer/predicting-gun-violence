# Data Processing

In this section, I start with data preparation. I remove redundant columns to reduce the number of features. Next, I merge the  county health datasets together. I then aggregate columns from the gun violence dataset to the county level. Finally, I merge the county health and gun violence data together. 

Next, I move on to analysis and data transformation. I run the same analysis as above on the joined dataset to check skewness and missing values. Next I plot distributions of the columns to check if they are normally distributed. I also plot scatterplots to check for outliers. I then impute the missing values and transform the data.

### Impute unreliable values
A column is provided to flag 'unreliable' values for Low Birth Weight (LBW). Using this, I impute LBW with the mean.

```python
# Impute 'Unreliable' values of '% LBW' with mean; Once imputed, I will no longer need the 'Unreliable' (Y/N) column
county_health1_df.rename(columns={'% LBW':'Perc_LBW'}, inplace=True);
mean_LBW = county_health1_df.Perc_LBW.mean()
unreliable_LBW = county_health1_df.Unreliable == 'x';
county_health1_df.loc[unreliable_LBW, "Perc_LBW"] = mean_LBW;
```

### Remove unneeded columns

Each column provided in the county health data also has 95% High and Low Confidence Interval columns, which are redundant. I also remove socio-economic and environment factors, as I am focussed on health factors only.  I keep 'Uninsured' and remove 'Uninsured Adult' and 'Uninsured Child' as they are redundant.

```python
# Remove redundant columns, e.g. 95% CI Low and High, and non-health factors from health1 dataframe
county_health1_reduced = county_health1_df.iloc[:, [0,1,2,3,10,14,18,23,30,34,38,40,44,46,52,57,59,67,72,76,80,84,89,96]];

# Remove redundant columns, e.g. 95% CI Low and High, Uninsured Adult & Child as Uninsured (total) is in first tab, 
# and non-health factors from health2 dataframe
county_health2_reduced = county_health2_df.iloc[:, [0,4,11,18,24,27,30,34,36,38,40,41,43,46,57,58,77]];

# Check FIPS column is unique in both dataframes and can be used for joining
county_health2_reduced.FIPS.duplicated().sum()
county_health1_reduced.FIPS.duplicated().sum()
```

### Merge dataframes

Merge the two county health datasets together using the FIPS code.

```python
# Merge the two (reduced) County health datasets together using FIPS
county_health_all = pd.merge(county_health1_reduced, county_health2_reduced, on="FIPS")
county_health_all.head()
```

### Aggregate gun data

The gun data contains one row per gun incident. Since I want to analyse this for each county, I aggregate the number of gun deaths, injuries and incidents for each county first before merging this data with the county health data.

```python
# Group gun violence data by state and county (aggregate number killed, number injured and number incidents)
gun_violence_county_df = gun_violence_df.groupby(['derived_county_FIPS'], as_index=False).agg(
    {'n_killed': np.sum, 'n_injured': np.sum, 'incident_id': np.size})

gun_violence_county_df.rename(columns={'derived_county_FIPS':'FIPS', 'incident_id':'n_incidents'}, inplace=True)
```

### Merge dataframes

Next I merge the aggregated gun violence data with the county health data.

```python
# Merge gun violence data into county health dataframe
county_health_all = pd.merge(county_health_all, gun_violence_county_df, on=['FIPS'])
```

```python
# Check final cleaned and merged dataframe 
county_health_all.head()
```

### Create new features using percentages

Because I want to compare gun deaths, gun injuries and gun incidents by county, it makes sense to create new features for these as percentages per county population.

```python
# Generate new fields for percent of killed / injured / incidents per county population
county_health_all['perc_killed'] = (county_health_all.n_killed / county_health_all.Population)*100
county_health_all['perc_injured'] =  (county_health_all.n_injured / county_health_all.Population)*100
county_health_all['perc_incidents'] =  (county_health_all.n_incidents / county_health_all.Population)*100
```

# Model Build

In this section, I build multiple regression models using features selected from PCA and correlation rankings and compare both to find the best model.

### Model 1

A multiple regression model with the most important features from PCA above. I added features one at a time in order of importance until the R-squared no longer increased. This resulted in the top 8 features being used.

```python
features1 = county_health_all[['PCP Rate', 'Dentist Rate','% With Access','Preventable Hosp. Rate','Teen Birth Rate',
                             'Child Mortality Rate','Age-Adjusted Mortality 10s','MV Mortality Rate']]
```

### Split data and fit model

I split the data into 10 folds and use cross-validation to check the Mean Squared Error and R-squared.

```python
# Split into test and training sets
X = features1
y = county_health_all.perc_incidents
kf = KFold(n_splits=10)
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]
    gun_health_model1 = sm.OLS(y_train,X_train).fit()
    print("Mean Squared Error: ", gun_health_model1.mse_total, "   R-squared: ", gun_health_model1.rsquared)
```

The results above are consistent over the 10 folds with errors around 0.017 and R-squared at 0.86. This shows the model explains 86% of the variance in the target.

### Prediction and test error

Next, I use Model 1 to predict test values, calculate the test error and visualise the actuals versus the predicted values.

```python
# Predict test values
y_prediction = gun_health_model1.predict(X_test)

# Calculate error between actual and prediction
error = (y_prediction - y_test)/y_test*100
avg_error = np.sum(error)/2238
print("Model1 has an average error of:", avg_error, "%")
```

```python
# Scatter plot of model1 test prediction
plt.rcParams['figure.figsize'] = (6,4);
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Model 1 Predicted vs Actual')
plt.scatter(y_test, y_prediction, s=30, linewidth='0');
```
